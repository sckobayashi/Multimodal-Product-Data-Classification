### Image/Text Data Visualisation

```python
from IPython.core.display import display, HTML
from PIL import Image
import base64
import io
import glob
```

```python
# This should be a helper function

data = pd.DataFrame()
img_list = glob.glob('images/image_train/*.jpg')
    
for i in range(5):
    try:
        filename = img_list[i]

        img = Image.open(filename)
        img_buffer = io.BytesIO()
        img.save(img_buffer, format="JPEG")
        imgStr = base64.b64encode(img_buffer.getvalue()).decode("utf-8") 

        data.loc[i, 'img'] = '<img src="data:image/jpeg;base64,{0:s}">'.format(imgStr)
        data.loc[i, 'imageid'] = filename.split('_')[-3]

    except: print('')
        
data = data.astype({'imageid': 'int64'})
```

```python
data = data.join(txt_feats.set_index('imageid'), on='imageid') 
```

```python
html_all = data.to_html(escape=False)
display(HTML(html_all))
```

* image sizes are not standardized
